# Why Do We Need NumPy?

Before NumPy, Python used lists to store data.

Problems with lists:
- Slow for large data
- Cannot perform mathematical operations easily
- Not efficient for numerical computations

Example: 
If we want to multiply all values by 2, we need loops.

NumPy solves this problem.

### Key Differences

| Feature | Python List | NumPy Array |
|---|---|---|
| Speed | Slow | Very Fast |
| Data Type | Mixed allowed | Same type |
| Math Operations | Not optimized | Vectorized |
| Used in ML | Rarely | Always |

Example:

- First 2 rows: `arr[:2, :]`

````python
import numpy as np

data = np.array([[100, 200, 300, 400, 500],
````

```python
# Task 1: Get element 800 (row 1, col 2 - remember 0-indexing!)
element = data[1, 2]
print('Task 1:', element)  # 800

# Task 2: Entire first row (index 0)
first_row = data[0, :]
# or simply: data[0]
print('Task 2:', first_row)  # [100 200 300 400 500]

# Task 3: Entire last column (index -1 or 4)
last_col = data[:, -1]

subarray = data[:2, :3]
print('Task 4:\n', subarray)
# [[100 200 300]

# Task 5: All elements > 1000
filtered = data[data > 1000]
print('Task 5:', filtered)
# [1100 1200 1300 1400 1500 1600 1700 1800 1900 2000]

# Task 6: Middle 2×2 block (rows 1-2, cols 1-2)
middle = data[1:3, 1:3]
print('Task 6:\n', middle)
# [[ 700  800]
#  [1200 1300]]

# Task 7: Every other element from second row (index 1)
every_other = data[1, ::2]
print('Task 7:', every_other)  # [600 800 1000]

# Task 8: All elements in third column (index 2)
third_col = data[:, 2]
print('Task 8:', third_col)  # [ 300  800 1300 1800]

# Task 9: Replace values > 1500 with 9999
data_copy = data.copy()
data_copy[data_copy > 1500] = 9999
print('Task 9:\n', data_copy)
# [[ 100  200  300  400  500]
#  [ 600  700  800  900 1000]
#  [1100 1200 1300 1400 1500]
#  [9999 9999 9999 9999 9999]]

# Task 10: Bottom-right 2×2 corner
corner = data[-2:, -2:]
# or: data[2:4, 3:5]
print('Task 10:\n', corner)
# [[1400 1500]
#  [1900 2000]]
```

**Key Insights:**
- **Negative indexing** is powerful: `-1` = last, `-2` = second to last
- **Boolean indexing** returns a 1D array of matching elements
- **Slicing** preserves dimensions: `arr[1:3, 1:3]` gives 2D result
- **Step slicing** `[::2]` means "every other element"
- Always use `.copy()` when modifying to preserve original!

**Why Broadcasting is Powerful:**
1. **No explicit loops** - Code is clean and fast
2. **Memory efficient** - Smaller array isn't duplicated

**Broadcasting in Machine Learning:**
- **Normalization**: Scale features to same range (z-score)
- **Batch operations**: Apply operations to multiple samples at once
- **Neural networks**: Add bias to each neuron in a layer
- **Image processing**: Apply filters to all pixels
