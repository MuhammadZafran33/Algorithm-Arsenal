### List Comprehension:
###### 1. A concise way to create lists in python 2. Compact and easier to read then Traditional Loops..   3. Expression for Value in Iterable if condtion...


```python
doubles = []
for x in range(1, 11):
    doubles.append(x*2)
print(doubles)
```

```python
doubles = [x*2 for x in range(1, 11)]

print(doubles)
```

```python
doubles = [x*2 for x in range(1, 11)]
triples = [y*3 for y in range(1, 11)]
print(doubles)
print(triples)
```

```python
doubles = [x*2 for x in range(1, 11)]
triples = [y*3 for y in range(1, 11)]
squares = [z*z for z in range(1, 11)]
print(doubles)
print(triples)
print(squares)
```

```python
fruits = ["apple", "orange", "banana", "coconut"]
fruits = [fruit.upper() for fruit in fruits]
print(fruits)
```

```python

fruits = [fruit.upper() for fruit in ["apple", "orange", "banana", "coconut"]]
print(fruits)
```

```python
fruits = ["apple", "orange", "banana", "coconut"]
fruit_chars = [fruit[0] for fruit in fruits]
print(fruit_chars)
```

```python
numbers  = [1, -2, 3, -4, 5, -6, -7, 8]
positive_nums = [num for num in numbers if num >= 0]
negative_nums = [num for num in numbers if num < 0]
even_nums = [num for num in numbers if num % 2 == 0]
odd_nums = [num for num in numbers if num % 2 == 1]
print(positive_nums)
print(negative_nums)
print(even_nums)
print(odd_nums)
```

```python
grades = [85, 42, 56, 76, 98, 30]

passing_grades = [grade for grade in grades if grade>=60]
print(passing_grades)
```
