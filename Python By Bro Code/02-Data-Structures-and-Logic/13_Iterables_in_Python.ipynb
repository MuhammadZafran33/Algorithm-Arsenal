### Iterables:
###### An object/collection that can return its elements one at a time allowing it to be iterated over in Loop... e.g: Lists, sets, dict, tuples

```python
numbers = [1,2,3,4,5]
for number in numbers:
    print(number)
```

```python
numbers = [1,2,3,4,5]
for number in reversed(numbers):
    print(number, end=" - ")
```

```python
numbers = (1,2,3,4,5)
for number in numbers:
    print(number)
```

```python
fruites = {"apple", "orange", "banana", "cocnut"}

for fruit in fruites:
    print(fruit)
```

```python
fruites = {"apple", "orange", "banana", "cocnut"}

for fruit in reversed(fruites):
    print(fruit)
```

```python
name = "Muhammad zafran"

for characters in name:
    print(characters, end=" ")
```

```python
my_dictionary= {"A": 1, "B": 2, "C":3}

for key in my_dictionary:
    print(key)
```

```python
my_dictionary= {"A": 1, "B": 2, "C":3}

for value in my_dictionary.values():
    print(value)
```

```python
my_dictionary= {"A": 1, "B": 2, "C":3}

for key,value in my_dictionary.items():
    # print(key,value)
    print(f"{key} = {value}")
```
