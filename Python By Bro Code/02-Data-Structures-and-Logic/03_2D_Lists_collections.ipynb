```python
fruits = ["apple", "orange", "banana", "coconut"]
vegetables = ["celery", "carrots", "potatos"]
meats = ["chicken", "fish", "turkey"]

groceries = [fruits, vegetables, meats]

print(groceries)
print(groceries[0])
print(groceries[1])
print(groceries[2])
```

```python
print(groceries[0][2])
print(groceries[1][1])
print(groceries[2][0])
```

```python
groceries = [['apple', 'orange', 'banana', 'coconut'], 
['celery', 'carrots', 'potatos'], 
['chicken', 'fish', 'turkey']]
print(groceries)

for collection in groceries:
    for food in collection:
        print(food, end=" ")
```

```python
groceries = (('apple', 'orange', 'banana', 'coconut'), 
('celery', 'carrots', 'potatos'), 
('chicken', 'fish', 'turkey'))
print(groceries)

for collection in groceries:
    for food in collection:
        print(food, end=" ")
```

```python
groceries = [
    {'apple', 'orange', 'banana', 'coconut'}, 
{'celery', 'carrots', 'potatos'}, 
{'chicken', 'fish', 'turkey'}]
print(groceries)

for collection in groceries:
    for food in collection:
        print(food, end=" ")
```

```python
num_pad = ((1,2,3),
(4,5,6),
(7,8,9),
("*", 0, "#"))

for row in num_pad:
    for num in row:
        print(num, end=" ")
    print()
```
