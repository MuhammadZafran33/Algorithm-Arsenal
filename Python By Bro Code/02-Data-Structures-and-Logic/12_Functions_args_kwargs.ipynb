### *Args =
###### allow you to pass multiple non-key arguments...
### **kwargs = 
###### allow you to pass multiple keyword arguments * unpacking operator
###### 1. positional 2. default 3. keyword 4. arbitery


```python
def add(a, b):
    return a+b
 add(1,2)
# add(1,2,3)
```

```python
def add(*args):
    print(type(args))

add(1,2,3)
```

```python
def add(*args):
    total = 0
    for arg in args:
        total+=arg
    return total

print(add(1,2,3, 4,5))
print(add(1))
```

```python
def display_name(*args):
    for arg in args:
        print(arg, end=" ")
display_name("Mr", "Muhammad", "Zafran", "AI", "2nd year")
```

#### Key-word Arguments...

```python
def print_address(**kwargs):
    print(type(kwargs))

print_address(street="27",
            sector="D4",
            city="Hayatabad Peshawar",
            state="Pakistan",
            zip="25100")
```

```python
def print_address(**kwargs):
    for key in kwargs.keys():
        print(key)

print_address(street="27",
            sector="D4",
            city="Hayatabad Peshawar",
            state="Pakistan",
            zip="25100")
```

```python
def print_address(**kwargs):
    for value in kwargs.values():
        print(value)

print_address(street="27",
            sector="D4",
            city="Hayatabad Peshawar",
            state="Pakistan",
            zip="25100")
```

```python
def print_address(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_address(street="27",
            sector="D4",
            city="Hayatabad Peshawar",
            state="Pakistan",
            zip="25100")
```

##### Both *args and **kwargs

```python
def shipping_label(*args, **kwargs):
    for arg in args:
        print(arg, end=" ")
    print()
    for value in kwargs.values():
        print(value, end=" ")
shipping_label("Dr.", "Muhammad", "Zafran", "AI",
                street="27",
                sector="D4",
                city="Hayatabad Peshawar",
                state="Pakistan",
                zip="25100")
```
