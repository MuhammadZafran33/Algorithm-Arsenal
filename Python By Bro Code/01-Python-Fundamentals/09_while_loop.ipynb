```python
# While Loop = execute some code WHILE some condition remain True....
```

```python
name  = input("Enter your name: ")

while name == "":
    print("You didnot enter your name")
    name  = input("Enter your name: ")
print(f"Hello {name}")
```

```python
age  = int(input("Enter your Age: "))

while age < 0:
    print("Age cannot be negative Bro")
    age  = int(input("Enter your Age: "))
print(f"Hello you are {age} years old")
```

```python
food = input("Enter a food you like (q to quite)")

while not food == "q":
    print(f"You like {food}")
    food = input("Enter a food you like (q to quite)")

print("Bye Bye")
```

```python
num = int(input("Enter a number between 1-10: "))

while num<1 or num>10:
    print(f"{num} is not valid!")
    num = int(input("Enter a number between 1-10: "))

print(f"Your number is {num}")
```
